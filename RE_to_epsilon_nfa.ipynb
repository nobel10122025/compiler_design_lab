```
from pyformlang.finite_automaton import EpsilonNFA, State, Symbol, Epsilon
from pyformlang.regular_expression import Regex
import networkx as nx
```

```
x=input("enter a regular expression:")
regex = Regex(x)
enfa2 = regex.to_epsilon_nfa()
print("start state : {} ".format(enfa2.start_states))
print("final state : {} ".format(enfa2.final_states))
no_of_transition=enfa2.get_number_transitions()
print("number of transitions: {}".format(no_of_transition))
print(enfa2.states)
print(enfa2.to_dict())
graph=enfa2.to_networkx()
nx.draw_networkx(graph,node_size=500)
```

```
dfa=enfa2.to_deterministic()
graph_1=dfa.to_networkx()
nx.draw_networkx(graph_1)
print("The transition table:")
dfa.to_dict()
```
